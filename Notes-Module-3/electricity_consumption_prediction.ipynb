## 1.1 let's look at the non-positive values for each attr

```python
no_rec = df.shape[0]
for col_name in df.columns:
    ind_nonpos = df[col_name] <= 0
    print(col_name, ': ')
    for val in df[col_name][ind_nonpos].unique():
        prc = df[col_name][ind_nonpos].isin([val]).sum()/no_rec*100
        print(val, ' [', prc, ']')
```

```python
# ASSUMPTION zero and negative values have different meanings
# replace negative values with zero 
# and create the following encoding for each attribute
# attr_zero = 0 if attr_val = 0
# attr_neg = 1 if attr_val < 0
# These new attibutes will also be used in learning

def add_encodings(df):

    cols_interest = ['Coal_MW','Gas_MW','Hidroelectric_MW','Nuclear_MW','Wind_MW','Solar_MW','Biomass_MW']

    for col_name in cols_interest:
        col_zero = col_name + '_zero'
        col_neg = col_name + '_neg'
        
        df[col_zero] = df[col_name] == 0
        df[col_zero] = df[col_zero].astype(int)

        ind_tmp = df[col_name] < 0
        df[col_neg] = ind_tmp
        df[col_neg] = df[col_neg].astype(int)
        
        # whatever is negative set it to zero
        df.loc[ind_tmp, col_name] = 0
            
    return df

df = add_encodings(df)
df.columns
```

# 2. Adding some datetime related features

```python
def add_datetime_features(df):
    features = ["Year", "Week", "Day", "Dayofyear", "Month", "Dayofweek",
                "Is_year_end", "Is_year_start", "Is_month_end", "Is_month_start",
                "Hour", "Minute",]
    one_hot_features = ["Month", "Dayofweek"]

    datetime = pd.to_datetime(df.Date * (10 ** 9))

    df['Datetime'] = datetime  # We won't use this for training, but we'll remove it later

    for feature in features:
        new_column = getattr(datetime.dt, feature.lower())
        if feature in one_hot_features:
            df = pd.concat([df, pd.get_dummies(new_column, prefix=feature)], axis=1)
        else:
            df[feature] = new_column
    return df

df = add_datetime_features(df)
df['just_date'] = df['Datetime'].dt.date
print(df.columns)
print(df.shape)
```

# 3. Augument the dataset with temperature data

```python
df_temp = pd.read_csv("temp_OTP.csv")
df_temp['just_date'] = pd.to_datetime(df_temp['DATE']).dt.date
df_temp = df_temp.drop(columns=['DATE'])

print(df_temp.shape)
df_temp.head()
```

```python
# merge the two datasets
df = df.merge(df_temp, how='left', on='just_date')

# impute missing temperatures
df['MAX'] = df['MAX'].interpolate()
df['MIN'] = df['MIN'].interpolate()
df['TEMP'] = df['TEMP'].interpolate()
print(df.columns)
```

# 4. Split data into train / validation (leaving the last six months for validation)

```python
eval_from = df['Datetime'].max() + relativedelta(months=-6)  # Here we set the 6 months threshold
train_df = df[df['Datetime'] < eval_from].copy()
valid_df = df[df['Datetime'] >= eval_from].copy()

print(f"Train data: {train_df['Datetime'].min()} -> {train_df['Datetime'].max()} | {len(train_df)} samples.")
print(f"Valid data: {valid_df['Datetime'].min()} -> {valid_df['Datetime'].max()} | {len(valid_df)} samples.")
```

# 5. Data Reduction and Scalling

```python
label_col = "Consumption_MW"  # The target values are in this column

# Columns we do not need for training
to_drop = [label_col, "Date", "Datetime", "Production_MW", 'STATION', 'just_date']  

# train
y_train = train_df[label_col] 

df_train = train_df.drop(to_drop, axis=1)
X_train = df_train.values

train_columns = df_train.columns
print(train_columns)

use_pca = False

# perform PCA
if use_pca:
    pca = decomposition.PCA(n_components=20)
    X_train_pca = pca.fit_transform(X_train)
    print('PCA explain var = ', pca.explained_variance_ratio_.sum())

# scale train dataset
if use_pca:
    scaler = preprocessing.StandardScaler().fit(X_train_pca)
    X_train_scaled = scaler.transform(X_train_pca)
else:
    scaler = preprocessing.StandardScaler().fit(X_train)
    X_train_scaled = scaler.transform(X_train)

# validation dataset
y_valid = valid_df[label_col] 
X_valid = valid_df.drop(to_drop, axis=1).values
if use_pca:
    X_valid_pca = pca.transform(X_valid)
    X_valid_scaled = scaler.transform(X_valid_pca)
else:
    X_valid_scaled = scaler.transform(X_valid)
    
print(X_train_scaled.shape)
```

# 6. Train using linear regression & XGBoost

```python
# Linear Regression with Cross-Validation
l2_cv = RidgeCV(cv=None, alphas=np.linspace(0.0001,0.01,100), store_cv_values=True)
l2_cv.fit(X_train_scaled, y_train)
print('alpha = ', l2_cv.alpha_) 
print('coef = ', l2_cv.coef_)

print('L2 - expected RMSE = ',np.min(np.sqrt(l2_cv.cv_values_.mean(axis=0))))
```

```python
# XGBoost
xg_trn_data = xgb.DMatrix(X_train_scaled, label=y_train)
xg_vld_data = xgb.DMatrix(X_valid_scaled, label=y_valid)

num_round = 100
xgb_param = {"objective": "reg:squarederror" if xgb.__version__ > '0.82' else 'reg:linear',
            'eta': 0.15, 'booster': 'gbtree', 'max_depth': 5}
watchlist = [(xg_trn_data, "train"), (xg_vld_data, "valid")]

bst = xgb.train(xgb_param, xg_trn_data, num_round, watchlist)
```

# 7. Model evaluation

```python
all_pred = [l2_cv, bst]
all_names = ['L2', 'BST']
zip_pred = zip(all_names, all_pred)

X_train_new = np.nan
X_valid_new = np.nan

for (tmp_name, tmp_pred) in zip_pred:
    if tmp_pred.__class__ == bst.__class__:
        y_pred_val = tmp_pred.predict(xgb.DMatrix(X_valid_scaled))
        y_pred_train = tmp_pred.predict(xgb.DMatrix(X_train_scaled))
    else:
        y_pred_val = tmp_pred.predict(X_valid_scaled)
        y_pred_train = tmp_pred.predict(X_train_scaled)
    print(tmp_name, ': ', np.sqrt(mean_squared_error(y_valid, y_pred_val)))
    
    if np.isnan(X_valid_new).any():
        X_valid_new = y_pred_val.reshape([-1,1])
        X_train_new = y_pred_train.reshape([-1,1])   
    else:
        X_valid_new = np.hstack([X_valid_new, y_pred_val.reshape([-1,1])])
        X_train_new = np.hstack([X_train_new, y_pred_train.reshape([-1,1])])
```

```python
## 7. Prepare a Kaggle submission on test data.

my_predictor = bst

test_df = pd.read_csv("test_electricity.csv")
test_df = add_encodings(test_df)
test_df = add_datetime_features(test_df)
test_df['just_date'] = test_df['Datetime'].dt.date
test_df = test_df.merge(df_temp, how='left', on='just_date')
test_df['MAX'] = test_df['MAX'].interpolate()
test_df['MIN'] = test_df['MIN'].interpolate()
test_df['TEMP'] = test_df['TEMP'].interpolate()

df_test = test_df.drop(to_drop[1:], axis=1)
X_test = df_test.values

test_columns = df_test.columns
assert np.all(train_columns == test_columns), 'The columns are not in the same order'

# scale test dataset
if use_pca:
    X_test_pca =  pca.transform(X_test)
    X_test_scaled = scaler.transform(X_test_pca)
else:
    X_test_scaled = scaler.transform(X_test)

# predict and save
solution_df = pd.DataFrame(test_df["Date"])
if my_predictor.__class__ == bst.__class__:
    solution_df["Consumption_MW"] = my_predictor.predict(xgb.DMatrix(X_test_scaled))
else:
    solution_df["Consumption_MW"] = my_predictor.predict(X_test_scaled)

# save to file
solution_df.to_csv("submission_Kaggle.csv", index=False)
print("Done!")
```

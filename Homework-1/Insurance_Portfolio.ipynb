```python
# 1 Create a pandas data frame, df1, using the data in the ‘example_portfolio.csv’ file. Make sure that 'PolicyNo' is the index of this dataframe.
df = pd.read_csv('example_portfolio.csv') 
df = df.set_index('PolicyNo')
```

```python
# 4 Create a new dataframe, df2, that contains only the following fields 
# 'Vehicle_Anti_Theft_Device', 'Driver_Minimum_Age', 'Vehicle_Age_In_Years', 'Annual_Premium', 'Loss_Amount'. 
# Note that 'PolicyNo' is the index of this dataframe as well. 
# To answer the following questions use df2.

columns = df[['Vehicle_Anti_Theft_Device', 'Driver_Minimum_Age', 'Vehicle_Age_In_Years', 'Annual_Premium', 'Loss_Amount']] #get columns from first dataframe
df2 = columns.copy() # copy to second dataframe
df2
```

```python
# 5 Show the mean, std, min, and max for the numerical fields
print('MEAN:\n',    df2.mean())
print('\nSTD:\n',   df2.std())
print('\nMIN:\n',   df2.min())
print('\nMAX:\n',   df2.max())
```

```python
# 6 Show the unique values of 'Vehicle_Anti_Theft_Device'.
print('UNIQUE VALUES OF Vehicle_Anti_Theft_Device:')
for value in df2.Vehicle_Anti_Theft_Device.unique(): #neatly prints unique values from numpy array
    print('\t', value)
```

```python
# 7 Calculate and print out the Portfolio Premium, Portfolio Loss, and Loss Ratio. 
# Portfolio Premium = sum of all the Annual Premiums of all the policies. 
# Portfolio Loss = sum of all the Loss Amounts of all the policies. 
# Loss Ratio = Portfolio Loss / Portfolio Premium.

portfolio_premium = df2['Annual_Premium'].sum() #Portfolio Premium

portfolio_loss = df2['Loss_Amount'].sum() #Portfolio Loss

loss_ratio = portfolio_loss/portfolio_premium #Loss ratio

#Output results
print('PORTFOLIO PREMIUM:', portfolio_premium)
print('PORTFOLIO LOSS:', portfolio_loss)
print('LOSS RATIO:', loss_ratio)
```
